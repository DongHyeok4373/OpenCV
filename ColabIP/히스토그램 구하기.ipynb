```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/lena256.jpg')
RGB_img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
R_img1,G_img1,B_img1 = cv2.split(RGB_img1)
plt.figure(figsize=(20,20))
plt.subplot(111), plt.imshow(RGB_img1)
plt.axis("off")
```

```python
hist = cv2.calcHist([R_img1], [0], None, [256], [0,255])
plt.subplot(111), plt.plot(hist, color ='r')
plt.xlim([0,256])
hist = cv2.calcHist([G_img1], [1], None, [256], [0,255])
plt.plot(hist,color ='g')
plt.xlim([0,256])
hist = cv2.calcHist([B_img1], [2], None, [256], [0,255])
plt.plot(hist,color ='b')
plt.xlim([0,255])
plt.show()
```
